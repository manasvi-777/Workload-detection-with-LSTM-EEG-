```python
mat_rating = scipy.io.loadmat('/kaggle/input/mental-cognitive-workload-eeg-data-stew-dataset/rating.mat')
workload_ratings = mat_rating['rating'].flatten()

print(f"Shape of Workload Ratings: {workload_ratings.shape}")
print(f"Unique Ratings: {np.unique(workload_ratings)}")
```

```python
 # Preprocessing parameters
# -----------------------------
fs = 256                  # sampling rate
l_freq, h_freq = 1, 40    # band-pass filter 1-40 Hz
notch_freq = 50           # line noise
epoch_length_sec = 2      # 2-second epochs
overlap = 0.5             # 50% overlap
n_channels, n_samples, n_subjects = eeg_data_raw.shape
```

```python
def preprocess_subject(eeg_subject, fs, l_freq, h_freq, notch_freq, epoch_length_sec, overlap):
    """
    Preprocess EEG signals for one subject.
    
    Input:
        eeg_subject : np.array of shape (n_channels, n_samples)
        fs          : Sampling frequency (Hz), e.g. 128 for STEW dataset
        l_freq      : Low cut frequency for band-pass filter
        h_freq      : High cut frequency for band-pass filter
        notch_freq  : Frequency to notch out (e.g. 50 Hz or 60 Hz line noise)
        epoch_length_sec : Epoch length in seconds (e.g. 2 or 4)
        overlap     : Fractional overlap between epochs (e.g. 0.5 = 50%)
    
    Output:
        epochs : np.array of shape (n_epochs, n_channels, n_samples_per_epoch)
    """
    
    # 1. Wrap numpy array into MNE Raw object so we can use MNE’s filtering/ICA functions
    info = mne.create_info(ch_names=[f"Ch{i+1}" for i in range(eeg_subject.shape[0])],
                           sfreq=fs, ch_types='eeg')
    raw = mne.io.RawArray(eeg_subject, info)
    
    # 2. Band-pass filter: keep only brain-relevant frequencies (e.g. 1–45 Hz)
    #    Removes very slow drifts (<1 Hz) and high-frequency muscle noise (>45 Hz)
    raw.filter(l_freq, h_freq, fir_design='firwin')
    
    # 3. Notch filter: remove 50/60 Hz line noise from power supply
    #Removes powerline interference (50 Hz in EU/India, 60 Hz in US/Japan). 
    #This noise is always present because EEG amplifiers pick up AC mains electricity.
    #Without this step, the ML model may learn to classify noise instead of EEG signals.
    raw.notch_filter(notch_freq)
    
    # 4. Artifact removal using ICA:
    #    Separates EEG into components (eye blinks, heartbeat, muscle artifacts).
    #    In practice, we would identify and exclude bad components before applying.
    ica = mne.preprocessing.ICA(n_components=min(15, eeg_subject.shape[0]), random_state=97)
    ica.fit(raw)
    ica.exclude = []  
    raw = ica.apply(raw)
    
    # 5. Re-reference EEG: subtract average across all channels
    #    This balances activity across electrodes and reduces bias from poor reference electrode
    raw.set_eeg_reference('average', projection=False)
    
    # 6. Z-score normalization: standardize each channel (mean=0, std=1)
    #    Ensures all channels contribute equally to model training
    data = raw.get_data()
    data = (data - data.mean(axis=1, keepdims=True)) / data.std(axis=1, keepdims=True)
    
    # 7. Epoching: split continuous EEG into overlapping fixed-size windows
    #    Each epoch is (n_channels, n_samples_per_epoch) and will get one label
    samples_per_epoch = int(epoch_length_sec * fs)        # number of samples per epoch
    step = int(samples_per_epoch * (1 - overlap))         # step size for overlap
    epochs = [data[:, start:start+samples_per_epoch] 
              for start in range(0, data.shape[1]-samples_per_epoch+1, step)]
    
    # 8. Return all epochs stacked into one array
    #    Final shape: (n_epochs, n_channels, n_samples_per_epoch)
    return np.array(epochs)
```

```python
# Define the class mapping function first
def rating_to_class(rating):
    if 4 <= rating <= 5:
        return 0  # Normal
    elif 6 <= rating <= 7:
        return 1  # Moderate
    elif 8 <= rating <= 9:
        return 2  # High
    else:
        return None
```

```python
# Preprocess all subjects
# -----------------------------
all_subject_epochs = []
all_labels = []

for subj_idx in range(n_subjects): # Process the EEG data for one subject 
    eeg_subj = eeg_data_raw[:, :, subj_idx] 
    epochs = preprocess_subject(eeg_subj, fs, l_freq, h_freq, notch_freq, epoch_length_sec, overlap) 
    all_subject_epochs.append(epochs) 
    # Get the single rating for this subject 
    rating = workload_ratings[subj_idx] 
    # Convert it to a class (0, 1, or 2) 
    label = rating_to_class(rating) 
    # Create an array with that label repeated for every epoch from this subject 
    labels_for_subject = np.repeat(label, len(epochs)) 
    all_labels.append(labels_for_subject)
```

```python


class_labels = [rating_to_class(r) for r in workload_ratings]
print(class_labels)
```

```python
#Concatenate all epochs and labels into final NumPy arrays 
X = np.concatenate(all_subject_epochs, axis=0) 
y = np.concatenate(all_labels, axis=0)
```

```python
# Current shape: (epochs, channels, samples)
X = np.transpose(X, (0, 2, 1))  # now (epochs, samples, channels)
y = y
print(X.shape, y.shape)
```

LSTM

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset, random_split
```

```python
# ===============================
# 1. Convert data to PyTorch tensors
# ===============================
X_tensor = torch.tensor(X, dtype=torch.float32)  # shape: (epochs, time, channels)
y_tensor = torch.tensor(y, dtype=torch.long)             # shape: (epochs,)
```

```python
# ===============================
# 2. Train/Test split (90/10)
# ===============================
total_samples = X_tensor.shape[0]
train_size = int(0.9 * total_samples)
test_size = total_samples - train_size

dataset = TensorDataset(X_tensor, y_tensor)
train_dataset, test_dataset = random_split(dataset, [train_size, test_size])

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python

# ===============================
# 3. Define LSTM Model
# ===============================
class EEG_LSTM(nn.Module):
    def __init__(self, input_size=14, hidden_size=64, num_layers=2, num_classes=3):
        super(EEG_LSTM, self).__init__()
        self.lstm = nn.LSTM(input_size=input_size, hidden_size=hidden_size, 
                            num_layers=num_layers, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(hidden_size*2, num_classes)  # *2 for bidirectional
    
    def forward(self, x):
        # x: (batch, time, channels)
        out, _ = self.lstm(x)             # out: (batch, time, hidden*2)
        out = out[:, -1, :]               # Take last timestep
        out = self.fc(out)                # (batch, num_classes)
        return out

model = EEG_LSTM()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
```

```python
# ===============================
# 4. Define loss and optimizer
# ===============================
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python

# ===============================
# 5. Training loop
# ===============================
num_epochs = 100

for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    for X_batch, y_batch in train_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        
        optimizer.zero_grad()
        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
    
    avg_loss = total_loss / len(train_loader)
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}")
```

```python
import torch
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Make sure model is in evaluation mode
model.eval()

# Convert test data to torch tensors
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)  # shape: (num_samples, seq_len, channels)
y_test_tensor = torch.tensor(y_test, dtype=torch.long)     # shape: (num_samples,)

# Move to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
X_test_tensor = X_test_tensor.to(device)
y_test_tensor = y_test_tensor.to(device)

# Disable gradient calculation for evaluation
with torch.no_grad():
    outputs = model(X_test_tensor)           # Forward pass
    _, predicted = torch.max(outputs, 1)    # Get predicted class

# Move back to CPU for sklearn metrics
predicted = predicted.cpu().numpy()
y_true = y_test_tensor.cpu().numpy()

# Accuracy
acc = accuracy_score(y_true, predicted)
print(f"Test Accuracy: {acc*100:.2f}%\n")

# Detailed classification report
print("Classification Report:")
print(classification_report(y_true, predicted, digits=3))

# Confusion matrix
cm = confusion_matrix(y_true, predicted)
print("Confusion Matrix:")
print(cm)
```

```python
import torch
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
import numpy as np

# Convert test labels to one-hot (binarized)
n_classes = 3
y_true_bin = label_binarize(y_true, classes=[0,1,2])  # shape: (num_samples, n_classes)

# Get model predictions as probabilities (softmax)
model.eval()
with torch.no_grad():
    outputs = model(X_test_tensor)          # shape: (num_samples, n_classes)
    probs = torch.softmax(outputs, dim=1).cpu().numpy()  # move to CPU

# Compute ROC and AUC for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], probs[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot ROC curves
plt.figure(figsize=(8,6))
colors = ['blue', 'green', 'red']
class_names = ['Normal','Moderate','High']

for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label=f'ROC curve of class {class_names[i]} (AUC = {roc_auc[i]:0.2f})')

plt.plot([0,1], [0,1], 'k--', lw=2)
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Multi-class ROC Curve")
plt.legend(loc="lower right")
plt.show()
```

```python
# Early Stopping parameters
patience = 25
best_val_loss = float('inf')
epochs_no_improve = 0
n_epochs = 100

for epoch in range(n_epochs):
    # ====== Training ======
    model.train()
    train_loss = 0.0
    for X_batch, y_batch in train_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)

        optimizer.zero_grad()
        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()
        
        train_loss += loss.item() * X_batch.size(0)
    
    train_loss /= len(train_loader.dataset)

    # ====== Validation ======
    model.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        for X_batch, y_batch in test_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            outputs = model(X_batch)
            loss = criterion(outputs, y_batch)
            val_loss += loss.item() * X_batch.size(0)

            _, predicted = torch.max(outputs, 1)
            total += y_batch.size(0)
            correct += (predicted == y_batch).sum().item()
    
    val_loss /= len(test_loader.dataset)
    val_acc = correct / total * 100

    print(f"Epoch [{epoch+1}/{n_epochs}] "
          f"Train Loss: {train_loss:.4f} "
          f"Val Loss: {val_loss:.4f} "
          f"Val Acc: {val_acc:.2f}%")
    
    # ====== Early Stopping Check ======
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        epochs_no_improve = 0
        torch.save(model.state_dict(), "best_model.pt")  # Save best model
    else:
        epochs_no_improve += 1
        if epochs_no_improve >= patience:
            print(f"Early stopping triggered after {epoch+1} epochs.")
            break

# Load best model after training
model.load_state_dict(torch.load("best_model.pt"))
```
